## Handling missing age values on AGE column

```python
mean_age= df["Age"].mean()
mean_age
```

```python
std_age = df["Age"].std()
std_age
```

```python
low_age = mean_age - std_age
low_age
```

```python
high_age = mean_age + std_age
high_age
```

```python
random_ages = np.random.randint(low_age,high_age,263)
random_ages
```

```python
age_copy = df["Age"].copy()
age_copy
```

```python
age_copy[np.isnan(age_copy)] = random_ages #replaces nan values with random_ages
age_copy
```

```python
df["Age"] = age_copy
```

```python
df.isnull().sum()
```

# Model Buliding

```python
X = df.drop("Survived",axis = 1)
X
```

```python
Y = df["Survived"]
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.30,random_state = 42)
```

```python
model = LogisticRegression(max_iter = 500)
```

```python
model.fit(x_train,y_train)
```

```python
predicted_res = model.predict(x_test)
predicted_res
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,predicted_res) #tp+tn/tp+tn+fp+fn 
```

```python
confusion_matrix(y_test,predicted_res) # (tp,fp,tn,fn)
```

#### 218 -> 1 survived : model -> survived {TP}
#### 22-> 0 notsurvived : model -> survived {FP}
#### 41-> 0 notsurvived : model -> notsurvived {TN}
#### 112 -> 1 survived : model -> notsurvived {FN}

```python
print(classification_report(y_test,predicted_res))
```

```python
y_test.value_counts()
```
